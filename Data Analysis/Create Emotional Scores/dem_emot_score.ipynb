```python
#dem script for attaching negative valence scores to tweets
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from numpy import dot
from numpy.linalg import norm

INPUT = '/home/twalton_umass_edu/Political Polarization Project/tmls/word_embeddings/sent_embed/'
OUPUT = '/home/twalton_umass_edu/Political Polarization Project/tmls/word_embeddings/sent_embed/final/'

num_words = [50, 100, 150, 200]
```

```python
#cosine similarity function
def cosine_sim(a, b):
    return dot(a, b) / (norm(a)*norm(b))
```

```python
for i in range(8):
    #load cleaned text dataframe
    text = pd.read_csv(INPUT + 'cleaned/dem_clean_' + str(i) + '.csv')
    #load sentence embeddings
    embedding = pd.read_csv(INPUT + 'trained/dem_trained_' + str(i) + '.csv')
    
    print('length of text: ' + str(len(text)))
    print('length of embedding: ' + str(len(text)))
    
    #remove unnamed axxis from embedding
    embedding = embedding.drop(['Unnamed: 0'], axis = 1)
    
    for num in num_words:
        neg = np.load('/home/twalton_umass_edu/Political Polarization Project/tmls/word_embeddings/emotions/dem_neg_dimension_' + str(num) + '.npy')
        #vector for holding scores
        scores = []
    
        #loop through and calculate cosine similarity
        for tweet in range(len(text)):
            score = cosine_sim(embedding.loc[tweet], neg)
            scores.append(score)
            if tweet % 100000 == 0:
                print(tweet)
    
        #create new variable
        text['neg_emot_score_' + str(num)] = scores
    
    #save output
    text.to_csv(OUPUT + 'dem_final_' + str(i) + '.csv')
    
    print(i)
```
